## Dividir a coluna em várias colunas

```python
def divide_TagsColumn():
    DesirableTags = ["ID", "Parent", "biotype"]
    IDs = []
    Parents = []
    biotypes = []
    Other_tags = []
    for transcript_tags in tanscripts_df["Tags"]:    #for each row in Tags column
        otherTags = []
        if "ID=" not in str(transcript_tags):
            IDs.append(np.nan)
        if "Parent=" not in str(transcript_tags):
            Parents.append(np.nan)
        if "biotype=" not in str(transcript_tags):
            biotypes.append(np.nan)
        if ";" in str(transcript_tags):
            row_tags = transcript_tags.split(";")
            for tag in row_tags:
                Tag_type = tag.split("=")[0]
                if Tag_type == "ID":
                    IDs.append(tag)
                elif Tag_type == "Parent":
                    Parents.append(tag)
                elif Tag_type == "biotype":
                    biotypes.append(tag)
                else:
                    otherTags.append(tag)
        else:
            if "ID=" in str(transcript_tags):
                IDs.append(np.nan)
            elif "Parent=" in str(transcript_tags):
                Parents.append(np.nan)
            elif "biotype=" in str(transcript_tags):
                biotypes.append(np.nan)
            else:
                otherTags.append(str(transcript_tags))
        Other_tags.append(str(otherTags))
    print(len(IDs))
    print(len(Parents))
    print("Parent" in Other_tags)
    tanscripts_df.insert(len(tanscripts_df.columns), "ID", IDs, True)
    tanscripts_df.insert(len(tanscripts_df.columns), "Parent", Parents, True)
    tanscripts_df.insert(len(tanscripts_df.columns), "Biotype", biotypes, True)
    tanscripts_df.insert(len(tanscripts_df.columns), "Other_tags", str(Other_tags), True)
    tanscripts_df.drop(['Tags'], axis=1)
    return tanscripts_df
        
```

```python
def element_matrix(chromosomes_df, sequence_df):
    """
    The matrix representing all intances of a sequence type available from the original transcripts dataframe
    sequence_df = Genes, CDSs, 5 or 3 prime UTR, Exons
    """
    Chromosomes_matrix = []
    for line in chromosomes_df.index:
        chromosome = chromosomes_df["Chr"][line]
        #print("searching citobands in chromossome: ", chromosome)
        chr_size = int(chromosomes_df["Last_index"][line])
        chr_sequence = sequence_df.loc[sequence_df['Chr'] == chromosome]
        chr_array = np.zeros((chr_size), dtype=int)
        #elements region
        for row in chr_sequence.index:
            chr_array[int(chr_sequence["First_index"][row]) : int(chr_sequence["Last_index"][row])] += 1
                
        
        Chromosomes_matrix.append(chr_array)
    return Chromosomes_matrix
```

```python
### Função para considerar todo o intervalo entre o primeiro indice até ao ultimo indice dos transcritos
def element_chr_array(chromosomes_df, sequence_df, chr_n):
    """
    The matrix representing all intances first instance until the last instance of the sequence type from the original transcripts dataframe
    sequence_df = Genes, CDSs, 5 or 3 prime UTR, Exons
    chr_n: String only number ex: "1"
    """
    print("searching cytobands in chromossome: ", chr_n)
    chromosome = chromosomes_df.loc[chromosomes_df["Chr"] == chr_n]
    chr_size = int(chromosome["Last_index"])
    chr_sequence = sequence_df.loc[sequence_df['Chr'] == chr_n]
    chr_array = np.zeros((chr_size), dtype=int)
    #elements region
    sequence_indexes = []
    for row in chr_sequence.index:
        sequence_indexes.append(int(chr_sequence["First_index"][row]))
        sequence_indexes.append(int(chr_sequence["Last_index"][row]))
        sequence_indexes = [min(sequence_indexes), max(sequence_indexes)]
    print("element's initial and final indexes:", sequence_indexes)
    chr_array[sequence_indexes[0] : sequence_indexes[1]] += 1
    #print(chr_array[sequence_indexes[0]-20 : sequence_indexes[1]]+20)
    return chr_array


#Juntar os arrays talvez não é boa ideia, memória
```

```python
def calc_elementDensity_chr(cytobands_df, chr_array):
    #cytobands_df = cytobands_df.drop(cytobands_df[cytobands_df['Chromossome'] == 'chrM'].index)
    chr_densities = []
    chr_cytobands = cytobands_df.loc[cytobands_df['Chromossome'] == cytobands_df["Chromossome"].unique()[chrom]]
    #Calculo da densidade para cada cytoband
    for row in chr_cytobands.index:
        cytoband_array = chr_array[int(chr_cytobands["First_index"][row]):int(chr_cytobands["Last_index"][row])]
        dr_length = np.count_nonzero(cytoband_array)
        cytoband_length = chr_cytobands["Size"][row]
        density = dr_length/cytoband_length
        chr_densities.append(round(density, 10))
    return chr_densities   
    #cytobands_df.insert(len(cytobands_df.columns), col_name, densities, True)
```

```python
def calc_ExonIntronDensity_chr(cytobands_df, sequence_array, exons_array, chr_n):
    #cytobands_df = cytobands_df.drop(cytobands_df[cytobands_df['Chromossome'] == 'chrM'].index)
    intron_densities = []
    exon_densities = []
    chr_cytobands = cytobands_df.loc[cytobands_df['Chromossome'] == "Chr" + str(chr_n)]
    #Calculo da densidade para cada cytoband
    for row in chr_cytobands.index:
        region_cytoband_array = sequence_array[int(chr_cytobands["First_index"][row]):int(chr_cytobands["Last_index"][row])]
        exon_cytoband_array = exons_array[int(chr_cytobands["First_index"][row]):int(chr_cytobands["Last_index"][row])]
            
        region_indexes = np.where(region_cytoband_array != 0)[0].tolist()
        exon_indexes = np.where(exon_cytoband_array != 0)[0].tolist()
        exon_length = len(np.intersect1d(region_indexes, exon_indexes))    #indices que estam na região e são exões
        intron_length = len(np.setdiff1d(region_indexes, exon_indexes))      #indices que estam na região mas não são exões
            
        cytoband_length = chr_cytobands["Size"][row]
        exon_density = exon_length/cytoband_length
        intron_density = intron_length/cytoband_length
        exon_densities.append(round(exon_density, 10))
        intron_densities.append(round(intron_density, 10))   #max 6    
    return exon_densities, intron_densities
```

```python
def chr_process(chromosomes_df, chr_n, cytobands_df):
    #for chr in chromosomes_df["Chr"]:
    sequence_dfs = [CDSs, five_prime_UTR, three_prime_UTR, exons]
    sequence_types = ["CDS", "five_prime_UTR", "three_prime_UTR", "exon"]
    ### criar já o array do chromossoma correspondente à presença de exons
    chr_exons_array = element_chr_array(chromosomes_df, exons, chr_n)
    ### Para cada um dos 3 tipos de sequencias calcular as densidades necessárias
    ###CDS
    chr_cds_array = element_chr_array(chromosomes_df, cds_df, chr_n)
    chr_cds_dens = calc_elementDensity_chr(cytobands_df, chr_array)
    cds_exon_dens, cds_intron_dens = calc_ExonIntronDensity_chr(cytobands_df, chr_array, chr_exons_array)
    ###five_prime_UTR
    chr_utr5_array = element_chr_array(chromosomes_df, utr5_df, chr_n)
    chr_utr5_dens = calc_elementDensity_chr(cytobands_df, chr_array)
    utr5_exon_dens, utr5_intron_dens = calc_ExonIntronDensity_chr(cytobands_df, chr_array, chr_exons_array)
    ###three_prime_UTR
    chr_utr3_array = element_chr_array(chromosomes_df, utr3_df, chr_n)
    chr_utr3_dens = calc_elementDensity_chr(cytobands_df, chr_array)
    utr3_exon_dens, utr3_intron_dens = calc_ExonIntronDensity_chr(cytobands_df, chr_array, chr_exons_array)
    ###exon
    chr_exon_dens = calc_elementDensity_chr(cytobands_df, chr_exons_array)
    ###Guardar as variáveis em arrays separados dentro ou fora da função
    return chr_cds_dens, cds_exon_dens, cds_intron_dens, chr_utr5_dens, utr5_exon_dens, utr5_intron_dens, chr_utr3_dens, utr3_exon_dens, utr3_intron_dens, chr_exon_dens

    
```

Funções de calculo de densidade

```python
def calc_elementDensity(cytobands_df, region_matrix, col_name="Element_density"):
    #cytobands_df = cytobands_df.drop(cytobands_df[cytobands_df['Chromossome'] == 'chrM'].index)
    densities = []
    for chrom in range(len(cytobands_df["Chromossome"].unique())):             
        chr_cytobands = cytobands_df.loc[cytobands_df['Chromossome'] == cytobands_df["Chromossome"].unique()[chrom]]
        #Calculo da densidade para cada cytoband
        for row in chr_cytobands.index:
            cytoband_array = region_matrix[chrom][int(chr_cytobands["First_index"][row]):int(chr_cytobands["Last_index"][row])]
            dr_length = np.count_nonzero(cytoband_array)
            cytoband_length = chr_cytobands["Size"][row]
            density = dr_length/cytoband_length
            densities.append(round(density, 10))
        
    cytobands_df.insert(len(cytobands_df.columns), col_name, densities, True)
```

```python
def calc_ExonIntronDensity(cytobands_df, region_matrix, exons_matrix, col_name1="exon_density", col_name2="intron_density"):
    #cytobands_df = cytobands_df.drop(cytobands_df[cytobands_df['Chromossome'] == 'chrM'].index)
    intron_densities = []
    exon_densities = []
    for chrom in range(len(cytobands_df["Chromossome"].unique())):
        #print("Calculating densities in chromosome ", cytobands_df["Chromossome"].unique()[chrom], " ...")
        chr_cytobands = cytobands_df.loc[cytobands_df['Chromossome'] == cytobands_df["Chromossome"].unique()[chrom]]
        #Calculo da densidade para cada cytoband
        for row in chr_cytobands.index:
            ############################ Resolver problema, o conjunto dos exons contem todo o conjunto dos CDSs ###############################
            region_cytoband_array = region_matrix[chrom][int(chr_cytobands["First_index"][row]):int(chr_cytobands["Last_index"][row])]
            exon_cytoband_array = exons_matrix[chrom][int(chr_cytobands["First_index"][row]):int(chr_cytobands["Last_index"][row])]
            #print("cytoband: ", chr_cytobands["Citoband"][row])
            #print("array_regiao: ", region_cytoband_array)
            #print("exon_regiao: ", exon_cytoband_array)
            region_indexes = np.where(region_cytoband_array != 0)[0].tolist()
            exon_indexes = np.where(exon_cytoband_array != 0)[0].tolist()
            #print("number of elements of region: ", len(region_indexes))
            #print("number of exons in region: ", len(np.intersect1d(region_indexes, exon_indexes)))
            #print("exons in CDS region: ", np.intersect1d(region_indexes, exon_indexes))
            #print("introns in CDS region: ", np.setdiff1d(region_indexes, exon_indexes))
            exon_length = len(np.intersect1d(region_indexes, exon_indexes))    #indices que estam na região e são exões
            intron_length = len(np.setdiff1d(region_indexes, exon_indexes))      #indices que estam na região mas não são exões
            #for region_element in region_indexes:
            #    if region_element in exon_indexes:
            #        exon_length+=1
            #    elif region_element not in exon_indexes:
            #        intron_length+=1
            cytoband_length = chr_cytobands["Size"][row]
            exon_density = exon_length/cytoband_length
            intron_density = intron_length/cytoband_length
            exon_densities.append(round(exon_density, 10))
            intron_densities.append(round(intron_density, 10))   #max 6
            
        
    cytobands_df.insert(len(cytobands_df.columns), col_name1, exon_densities, True)
    cytobands_df.insert(len(cytobands_df.columns), col_name2, intron_densities, True)
```

```python
def writeNewtxt(file_path, df):
    """Writes a pandas dataframe into txt, columns separated by tabs ("\t")
    """
    f = open(file_path, 'w')
    df_cols = ""
    for col in df.columns[:-1]:
        df_cols += col + "\t"
    df_cols += df.columns.values[-1]+ "\n"
    f.write(str(df_cols))
    for row in df.values:
        df_row = ""
        for col in row[:-1]:
            df_row += str(col) + "\t"
        df_row += str(row[-1])+ "\n"
        f.write(df_row)
    f.close
```

```python
def study_transcriptDens(cytobands_df, element_type, tanscripts_df, chromosomes_df): #element_type = "CDS"
    #chromosomes_df = tanscripts_df.loc[tanscripts_df['Type'] == "chromosome"]
    sub_trascripts_df = tanscripts_df.loc[tanscripts_df['Type'] == element_type]
    print("building matriz...")
    transcripts_matrix = element_matrix(chromosomes_df, sub_trascripts_df)
    colname = element_type+"s_density"
    print("doing calculations...")
    calc_elementDensity(cytobands_df, transcripts_matrix, colname)
    writeNewtxt(cytoband_write, cytobands_df)
```

```python
def sumOf_columns(df, colums=["CDSIntron_density", "UTR5Intron_density", "UTR3Intron_density"], newColName = "Introns_density"):
    """
    colums is a list of string
    """
    #cytobands_df = cytobands_df.drop(cytobands_df[cytobands_df['Chromossome'] == 'chrM'].index)
    new_densites = []
    for row in df.index:
        density =0
        for column in colums:
            density += df[column][row]
        new_densites.append(round(density, 3))
        
    df.insert(len(df.columns), newColName, new_densites, True)
```
