```python
genes_list = []
for row in dg_df["attributes"]:
    row_elements = row.split(";")
    for i in range (1, len(row_elements)-1, 2):
        gene_symbol = row_elements[i].split("=")[1]
        genes_list.append(gene_symbol)

len(genes_list)
```

```python
def writeNewtxt(file_path):
    """Writes a pandas dataframe into txt, columns separated by tabs ("\t")
    """
    f = open(file_path, 'w')
    string = ""
    for gene_symbol in genes_list:
        string += " " + gene_symbol
    f.write(string)
    f.close
```

```python
def create_newCol():
    ENG_ids = []
    converted_line_n = 0
    for row in dg_df["attributes"]:
        ENG_ids_row = ""
        for i in range (row.count("gene=")):
            dg = idConvert_df["converted_alias"][converted_line_n]
            ENG_ids_row += str(dg) + ", "
            converted_line_n += 1
        ENG_ids.append(str(ENG_ids_row[:-2]))
    return ENG_ids
    
```

```python
def writeNewtxt(file_path, df):
    """Writes a pandas dataframe into txt, columns separated by tabs ("\t")
    """
    f = open(file_path, 'w')
    str_cols = ""
    for col in df.columns[:-1]:
        str_cols += col + "\t"
    str_cols += df.columns.values[-1]+ "\n"
    f.write(str(str_cols))
    for row in df.values:
        str_row = ""
        for col in row[:-1]:
            str_row += str(col) + "\t"
        str_row += str(row[-1])+ "\n"
        f.write(str_row)
    f.close
```

```python
# Nao é util para já agregar todos od conjuntos, mais apropriado fazer um mapa de cada coisa

def calc_exons_introns(chromosomes_df, sequence_df, exons_df, genes_df, regions_df):
    """interval_df = Genes or CDSs
    +2⁰ = 1: is exon
    +2¹ = 2: is part of a region
    +2² = 4: is part of CDS
    +2³ = 8: is part of gene
    """
    Chrmosomes_matrix = []
    for line in chromosomes_df.index:
        chromosome = chromosomes_df["Chr"][line]
        print("searching citobands in chromossome: ", chromosome)
        chr_size = int(chromosomes_df["Last_index"][line])
        chromosome_elements = sequence_df.loc[sequence_df['Chr'] == chromosome]
        chromosome_exons = exons_df.loc[exons_df['Chr'] == chromosome]
        chr_genes = genes_df.loc[genes_df['Chr'] == chromosome]
        chr_regions = regions_df.loc[regions_df['Chr'] == chromosome]
        element_introns = np.zeros((chr_size), dtype=int)
        #genes region
        for row in chr_genes.index:
            for i in range (int(chr_genes["First_index"][row]), int(chr_genes["Last_index"][row])):
                element_introns[i]+=8
        #CDS region
        for row in chromosome_elements.index:
            for i in range (int(chromosome_elements["First_index"][row]), int(chromosome_elements["Last_index"][row])):
                element_introns[i]+=4
        #other regions
        for row in chr_regions.index:
            for i in range (int(chr_regions["First_index"][row]), int(chr_regions["Last_index"][row])):
                element_introns[i]+=2
        #exons region
        for row in chromosome_exons.index:
            for i in range (int(chromosome_exons["First_index"][row]), int(chromosome_exons["Last_index"][row])):
                element_introns[i]+=1
        
        Chrmosomes_matrix.append(element_introns)
    return Chrmosomes_matrix
```

```python
def calc_group(chromosomes_df, sequence_df):
    """
    sequence_df = Genes or CDSs, exons or other_regions
    """
    Chromosomes_matrix = []
    for line in chromosomes_df.index:
        chromosome = chromosomes_df["Chr"][line]
        print("searching citobands in chromossome: ", chromosome)
        chr_size = int(chromosomes_df["Last_index"][line])
        chr_sequence = sequence_df.loc[sequence_df['Chr'] == chromosome]
        chr_array = np.zeros((chr_size), dtype=int)
        #elements region
        for row in chr_sequence.index:
            for i in range (int(chr_sequence["First_index"][row]), int(chr_sequence["Last_index"][row])):
                chr_array[i]+=1
        
        Chromosomes_matrix.append(chr_array)
    return Chromosomes_matrix
```

```python
def calc_introns(chromosomes_df, CDS_df, exons_df):
    """
    sequence_df = Genes or CDSs, exons or other_regions
    if CDS element: +2
    if exon: +1
    introns = 2
    exons = 3
    """
    Chromosomes_matrix = []
    for line in chromosomes_df.index:
        chromosome = chromosomes_df["Chr"][line]
        print("searching citobands in chromossome: ", chromosome)
        chr_size = int(chromosomes_df["Last_index"][line])
        chr_CDS = CDS_df.loc[CDS_df['Chr'] == chromosome]
        chromosome_exons = exons_df.loc[exons_df['Chr'] == chromosome]
        chr_array = np.zeros((chr_size), dtype=int)
        #elements region
        #CDS region
        for row in chr_CDS.index:
            for i in range (int(chr_CDS["First_index"][row]), int(chr_CDS["Last_index"][row])):
                chr_array[i]+=2
        #exons region
        for row in chromosome_exons.index:
            for i in range (int(chromosome_exons["First_index"][row]), int(chromosome_exons["Last_index"][row])):
                chr_array[i]+=1
        
        Chromosomes_matrix.append(chr_array)
    return Chromosomes_matrix
```
