```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
######################################## calc densities ############################################
def calc_DRDensity(cytobands_df, dr_df):
    densities = []
    for chrom in cytobands_df["Chromossome"].unique():             
        print("searching cytobands in chromossome: ", chrom)
        chr_cytobands = cytobands_df.loc[cytobands_df['Chromossome'] == chrom]
        chr_drs = dr_df.loc[dr_df['#chrom'] == chrom]
        chr_size = chr_cytobands["Last_index"][chr_cytobands["Last_index"].index[-1]]+1
        cytobands_dr = np.zeros((chr_size), dtype=int)
        #percorrer o dataset pelas vulnerabilidades no cromossoma e anotar na lista np.zeroes
        for row in chr_drs.index:
            for i in range (chr_drs["start"][row], chr_drs["end"][row]):
                cytobands_dr[i]+=1
        #Calculo da densidade para cada cytoband
        for index in chr_cytobands.index:
            cytoband_array = cytobands_dr[int(chr_cytobands["First_index"][index]):int(chr_cytobands["Last_index"][index])+1]
            dr_length = np.count_nonzero(cytoband_array)
            cytoband_length = chr_cytobands["Size"][index]
            density = dr_length/cytoband_length
            densities.append(round(density, 10))
        
    cytobands_df.insert(len(cytobands_df.columns), "DRs_density", densities, True)


def writeNewtxt(file_path, df):
    """Writes a pandas dataframe into txt, columns separated by tabs ("\t")
    """
    f = open(file_path, 'w')
    df_cols = ""
    for col in df.columns[:-1]:
        df_cols += col + "\t"
    df_cols += df.columns.values[-1]+ "\n"
    f.write(str(df_cols))
    for row in df.values:
        df_row = ""
        for col in row[:-1]:
            df_row += str(col) + "\t"
        df_row += str(row[-1])+ "\n"
        f.write(df_row)
    f.close
```

```python
cytoband_read = "/home/andre/Documents/trabalho/cytobands/cytobandFiltered_processed_str.txt"
cytoband_write = "/home/andre/Documents/trabalho/cytobands/cytobandFiltered_processed_str_dr.txt"
disease_regions_read = "/home/andre/Documents/trabalho/datasets/Disease_genes/genehancer/GeneHancer_v5.20_processed.gff"

cytobands_df = pd.read_csv(cytoband_read, sep="\t")#, header=None))
dr_df = pd.read_csv(disease_regions_read, sep="\t")

cytobands_df
```

```python
calc_DRDensity(cytobands_df, dr_df)
cytobands_df
```
