```python
def cal_intronLength1(main_onetranscript):
    """Calcular o comprimento de intrões, diferença do comprimento dos exões no CDS pelo comprimento do CDS"""
    exon_trnscrpt=main_onetranscript.loc[main_onetranscript4["Type"] == "exon"]
    cds_trnscrpt=main_onetranscript.loc[main_onetranscript4["Type"] == "CDS"]
    
    first_exon = exon_trnscrpt.index[0]
    first_cds = cds_trnscrpt.index[0]
    
    CDSexon_length = (float(exon_trnscrpt["Last_index"][first_exon])+1) - (float(cds_trnscrpt["First_index"][first_cds]))

    for exon_index in exon_trnscrpt.index[1:-1]:
        exon_lenth = (float(exon_trnscrpt["Last_index"][exon_index])+1) - (float(exon_trnscrpt["First_index"][exon_index])-1)
        CDSexon_length += exon_lenth

    last_exon = exon_trnscrpt.index[-1]
    last_cds = cds_trnscrpt.index[-1]

    CDSexon_length += (float(cds_trnscrpt["Last_index"][last_cds])) - (float(exon_trnscrpt["First_index"][last_exon])-1)
    cds_length = float(cds_trnscrpt["Last_index"][last_cds]) - float(cds_trnscrpt["First_index"][first_cds])
    CDSintron_length = cds_length - CDSexon_length
    return CDSintron_length
```

```python
def cal_intronLength2(main_onetranscript):#, cds_trnscrpt):
    """Calcular o comprimento de intrões, a soma dos espaços entre os exões num CDS ((end_exon1-1) - (start_exon2+1))"""
    exon_trnscrpt=main_onetranscript.loc[main_onetranscript4["Type"] == "exon"]
    intron_length = 0
    for index in range(len(exon_trnscrpt.index[:-1])):
        intron_length += float(exon_trnscrpt["First_index"][exon_trnscrpt.index[index+1]]) - float(exon_trnscrpt["Last_index"][exon_trnscrpt.index[index]]) -2
        
    return intron_length
    
```
