## Dividir a coluna Tags em várias colunas

```python
def divide_TagsColumn():
    DesirableTags = ["ID", "Parent"]#, "biotype"]
    IDs = []
    parents = []
    #biotypes = []
    names = []
    for transcript_tags in tanscripts_df["Tags"]:    #for each row in Tags column
        if ";" in str(transcript_tags):
            present_tags = []
            row_tags = transcript_tags.split(";")
            for tag in row_tags:
                Tag_type = tag.split("=")[0]
                if Tag_type == "ID":
                    IDs.append(tag)
                    present_tags.append("id")
                elif Tag_type == "Parent":
                    parents.append(tag)
                    present_tags.append("parent")
                elif Tag_type == "Name":
                    names.append(str(tag).split("=")[1])
                    present_tags.append("Name")
            if "id" not in present_tags:
                IDs.append(np.nan)
            if "parent" not in present_tags:
                parents.append(np.nan)
            if "Name" not in present_tags:
                names.append(np.nan)

        elif ";" not in str(transcript_tags):
            if "ID=" in str(transcript_tags):
                IDs.append(str(transcript_tags))
                parents.append(np.nan)
                names.append(np.nan)
            elif "Parent=" in str(transcript_tags):
                IDs.append(np.nan)
                parents.append(str(transcript_tags))
                names.append(np.nan)
            elif "Name=" in str(transcript_tags):
                IDs.append(np.nan)
                parents.append(np.nan)
                names.append(str(transcript_tags.split("=")[1]))
            else:
                IDs.append(np.nan)
                parents.append(np.nan)
                names.append(np.nan)

    tanscripts_df.insert(len(tanscripts_df.columns), "ID", IDs, True)
    tanscripts_df.insert(len(tanscripts_df.columns), "Parent", parents, True)
    tanscripts_df.insert(len(tanscripts_df.columns), "Name", names, True)
    tanscripts_df.drop(['Tags'], axis=1)
    return tanscripts_df
```

## Methods for calculating the densities

Uma vez que a representação dos cromossomas e da ocorrência dos elementos no cromossoma é feito cromossoma a cromossoma, o calculo tambem e feito da mesma cromossoma a cromossoma

```python
### Função para considerar todo o intervalo entre o primeiro indice até ao ultimo indice dos transcritos
def element_chr_array(chromosomes_df, sequence_df, chr_n):
    """
    An array all basePairs of a chromosome, the interval from first instance to the last instance of the sequence type
    sequence_df = Genes, CDSs, 5 or 3 prime UTR, Exons
    chr_n: String only number ex: "1"
    """
    #print("searching cytobands in chromossome: ", chr_n)
    chromosome_index = chromosomes_df.index[chromosomes_df["Chr"] == str(chr_n)]
    chr_size = int(chromosomes_df["Last_index"][chromosome_index].iloc[0])
    chr_sequence = sequence_df.loc[sequence_df['Chr'] == str(chr_n)]
    chr_array = np.zeros((chr_size), dtype=int)
    #elements region
    sequence_indexes = []
    for row in chr_sequence.index:
        sequence_indexes.append(int(float(chr_sequence["First_index"][row])))
        sequence_indexes.append(int(float(chr_sequence["Last_index"][row])))
        sequence_indexes = [min(sequence_indexes), max(sequence_indexes)]
    #print("element's initial and final indexes:", sequence_indexes)
    chr_array[sequence_indexes[0] : sequence_indexes[1]] += 1
    #print(chr_array[sequence_indexes[0]-20 : sequence_indexes[1]]+20)
    return chr_array

#Juntar os arrays talvez não seja boa ideia, memória


def element_array(chromosomes_df, sequence_df, chr_n):
    """
    An array representing all intances of a sequence type in a chromosome available from the original transcripts dataframe
    sequence_df = Genes, CDSs, 5 or 3 prime UTR, Exons
    """
    chromosome_index = chromosomes_df.index[chromosomes_df["Chr"] == str(chr_n)]
    chr_size = int(chromosomes_df["Last_index"][chromosome_index].iloc[0])
    chr_sequence = sequence_df.loc[sequence_df['Chr'] == str(chr_n)]
    chr_array = np.zeros((chr_size), dtype=int)
    #elements region
    for row in chr_sequence.index:
        chr_array[int(float(chr_sequence["First_index"][row])) : int(float(chr_sequence["Last_index"][row]))] += 1
    return chr_array
```

```python
def element_chr_trnscpt_array(chromosomes_df, main_trcpts_df, chr_n, element_type="CDS"):
    """
    The array representing all intances first instance until the last instance of the sequence type from the original transcripts dataframe
    main_trcpts_df = Genes, includes: CDSs, 5 or 3 prime UTR, Exons
    chr_n: String only number ex: "1"
    element_type can be: CDS, five_prime_UTR or three_prime_UTR
    """
    print("searching cytobands in chromossome: ", chr_n)
    chromosome = chromosomes_df.loc[chromosomes_df["Chr"] == chr_n]
    chr_size = int(chromosome["Last_index"].iloc[0])
    chr_array = np.zeros((chr_size), dtype=int)
    #elements region
    chr_trcpts = main_trcpts_df.loc[main_trcpts_df["Chr"] == chr_n]
    chr_transcripts = np.unique_all(chr_trcpts["Parent"].to_numpy())[0]
    for transcript in chr_transcripts:
        chr_onetrcpt = chr_trcpts.loc[chr_trcpts["Parent"] == str(transcript)]
        element_trnscrpt = chr_trcpts.loc[(chr_trcpts["Type"] == str(element_type)) & (chr_trcpts["Parent"] == str(transcript))]
        #print("element_trnscrpt: ", element_trnscrpt)
        first_element = element_trnscrpt.index[0]
        last_element = element_trnscrpt.index[-1]
        chr_array[int(float(element_trnscrpt["First_index"][first_element])) : int(float(element_trnscrpt["Last_index"][last_element]))] += 1
    
    #print(chr_array[sequence_indexes[0]-20 : sequence_indexes[1]]+20)
    return chr_array
```

```python
def calc_elementDensity_chr(cytobands_df, chr_array, chr_n):
    #cytobands_df = cytobands_df.drop(cytobands_df[cytobands_df['Chromossome'] == 'chrM'].index)
    chr_densities = []
    chr_cytobands = cytobands_df.loc[cytobands_df['Chromossome'] == "chr" + str(chr_n)]
    #print("cytobands of chr", chr_n, " indexes: ", chr_cytobands.index)
    #Calculo da densidade para cada cytoband
    for row in chr_cytobands.index:
        cytoband_array = chr_array[int(chr_cytobands["First_index"][row]):int(chr_cytobands["Last_index"][row])]
        dr_length = np.count_nonzero(cytoband_array)
        cytoband_length = chr_cytobands["Size"][row]
        density = dr_length/cytoband_length
        chr_densities.append(round(density, 10))
    return chr_densities   
    #cytobands_df.insert(len(cytobands_df.columns), col_name, densities, True)
```

```python
def calc_ExonIntronDensity_chr(cytobands_df, chr_sequence_array, chr_exons_array, chr_n):
    """In a given chromosome calculate de exon and intron densities within a given sequence type (e.g CDSs) for 
    every cytoband
    """
    #cytobands_df = cytobands_df.drop(cytobands_df[cytobands_df['Chromossome'] == 'chrM'].index)
    intron_densities = []
    exon_densities = []
    #print()
    chr_cytobands = cytobands_df.loc[cytobands_df['Chromossome'] == "chr" + str(chr_n)]
    #Calculo das densidades para cada cytoband
    for row in chr_cytobands.index:
        exon_length, intron_length = get_lengths(chr_cytobands, chr_sequence_array, chr_exons_array, row)
        cytoband_length = chr_cytobands["Size"][row]
        exon_density = exon_length/cytoband_length
        intron_density = intron_length/cytoband_length
        exon_densities.append(round(exon_density, 10))
        intron_densities.append(round(intron_density, 10))   #max 6    
    return exon_densities, intron_densities
```

```python
def get_lengths(chr_cytobands, chr_sequence_array, chr_exons_array, row):
    region_cytoband_array = chr_sequence_array[int(chr_cytobands["First_index"][row]):int(chr_cytobands["Last_index"][row])]
    exon_cytoband_array = chr_exons_array[int(chr_cytobands["First_index"][row]):int(chr_cytobands["Last_index"][row])]
    region_indexes = np.where(region_cytoband_array > 0)[0].tolist()
    #print("region_indexes: ", region_indexes)
    exon_indexes = np.where(exon_cytoband_array > 0)[0].tolist()
    #print("exon_indexes: ", exon_indexes)
    exon_length = len(np.intersect1d(region_indexes, exon_indexes))   #indices que estam na região e são exões
    intron_length = len(np.setdiff1d(region_indexes, exon_indexes))    #indices que estam na região mas não são exões
    return exon_length, intron_length
```

```python
###Não é possivel correr esta função de uma vez devido a restrições de memória
def chr_process(chromosomes_df, main_trnscpts_df, cytobands_df, chr_n):
    #for chr in chromosomes_df["Chr"]:
    cds_df = main_trnscpts_df.loc[(main_trnscpts_df["Type"] == "CDS") & (main_trnscpts_df["Chr"] == str(chr_n))]
    utr5_df = main_trnscpts_df.loc[(main_trnscpts_df["Type"] == "five_prime_UTR") & (main_trnscpts_df["Chr"] == str(chr_n))]
    utr3_df = main_trnscpts_df.loc[(main_trnscpts_df["Type"] == "three_prime_UTR") & (main_trnscpts_df["Chr"] == str(chr_n))]
    exons = main_trnscpts_df.loc[(main_trnscpts_df["Type"] == "exon") & (main_trnscpts_df["Chr"] == str(chr_n))]
    ### criar já o array do chromossoma correspondente à presença de exons
    chr_exons_array = element_chr_array(chromosomes_df, exons, chr_n)
    ### Para cada um dos 3 tipos de sequencias calcular as densidades necessárias
    ###CDS
    chr_cds_array = element_chr_array(chromosomes_df, cds_df, chr_n)
    chr_cds_dens = calc_elementDensity_chr(cytobands_df, chr_cds_array, chr_n)
    cds_exon_dens, cds_intron_dens = calc_ExonIntronDensity_chr(cytobands_df, chr_cds_array, chr_exons_array, chr_n)
    ###five_prime_UTR
    #print("5prime utr dataframe: ", utr5_df)
    chr_utr5_array = element_chr_array(chromosomes_df, utr5_df, chr_n)
    chr_utr5_dens = calc_elementDensity_chr(cytobands_df, chr_utr5_array, chr_n)
    utr5_exon_dens, utr5_intron_dens = calc_ExonIntronDensity_chr(cytobands_df, chr_utr5_array, chr_exons_array, chr_n)
    ###three_prime_UTR
    chr_utr3_array = element_chr_array(chromosomes_df, utr3_df, chr_n)
    chr_utr3_dens = calc_elementDensity_chr(cytobands_df, chr_utr3_array, chr_n)
    utr3_exon_dens, utr3_intron_dens = calc_ExonIntronDensity_chr(cytobands_df, chr_utr3_array, chr_exons_array, chr_n)
    ###exon
    chr_exon_dens = calc_elementDensity_chr(cytobands_df, chr_exons_array, chr_n)
    ###Guardar as variáveis em arrays separados dentro ou fora da função
    return chr_cds_dens, cds_exon_dens, cds_intron_dens, chr_utr5_dens, utr5_exon_dens, utr5_intron_dens, chr_utr3_dens, utr3_exon_dens, utr3_intron_dens, chr_exon_dens
```

```python
def chr_process1(chromosomes_df, main_trnscpts_df, cytobands_df, chr_exons_array, chr_n):
    cds_df = main_trnscpts_df.loc[(main_trnscpts_df["Type"] == "CDS") & (main_trnscpts_df["Chr"] == str(chr_n))]
    ###CDS
    chr_cds_array = element_chr_trnscpt_array(chromosomes_df, main_trnscpts_df, chr_n, "CDS")
    chr_cds_dens = calc_elementDensity_chr(cytobands_df, chr_cds_array, chr_n)
    cds_exon_dens, cds_intron_dens = calc_ExonIntronDensity_chr(cytobands_df, chr_cds_array, chr_exons_array, chr_n)
    return chr_cds_dens, cds_exon_dens, cds_intron_dens
    
def chr_process2(chromosomes_df, main_trnscpts_df, cytobands_df, chr_exons_array, chr_n):
    utr5_df = main_trnscpts_df.loc[(main_trnscpts_df["Type"] == "five_prime_UTR") & (main_trnscpts_df["Chr"] == str(chr_n))]
    ###five_prime_UTR
    chr_utr5_array = element_array(chromosomes_df, utr5_df, chr_n)
    chr_utr5_dens = calc_elementDensity_chr(cytobands_df, chr_utr5_array, chr_n)
    # It is assumed 5 prime UTR is only introns, because the region as no coding role
    #utr5_exon_dens, utr5_intron_dens = calc_ExonIntronDensity_chr(cytobands_df, chr_utr5_array, chr_exons_array, chr_n)
    return chr_utr5_dens#, utr5_exon_dens, utr5_intron_dens
    
def chr_process3(chromosomes_df, main_trnscpts_df, cytobands_df, chr_exons_array, chr_n):
    utr3_df = main_trnscpts_df.loc[(main_trnscpts_df["Type"] == "three_prime_UTR") & (main_trnscpts_df["Chr"] == str(chr_n))]
    ###three_prime_UTR
    chr_utr3_array = element_array(chromosomes_df, utr3_df, chr_n)
    chr_utr3_dens = calc_elementDensity_chr(cytobands_df, chr_utr3_array, chr_n)
    # It is assumed 3 prime UTR is only introns, because the region as no coding role
    #utr3_exon_dens, utr3_intron_dens = calc_ExonIntronDensity_chr(cytobands_df, chr_utr3_array, chr_exons_array, chr_n)
    return chr_utr3_dens#, utr3_exon_dens, utr3_intron_dens

def chr_process4(chromosomes_df, main_trnscpts_df, cytobands_df, chr_exons_array, chr_n):
    ###exons
    chr_exon_dens = calc_elementDensity_chr(cytobands_df, chr_exons_array, chr_n)
    return chr_exon_dens
```

```python
def run_allChr(chromosomes_df, main_trnscpts_df, cytobands_df):
    all_chrs_cdsDens = []
    all_chrs_cdsExonDens = []
    all_chrs_cdsIntronDens = []
    all_chrs_utr5Dens = []
    #all_chrs_utr5ExonDens = []
    #all_chrs_utr5IntronDens = []
    all_chrs_utr3Dens = []
    #all_chrs_utr3ExonDens = []
    #all_chrs_utr3IntronDens = []
    all_chrs_exon_dens = []
    #for chrom in range(len(cytobands_df["Chromossome"].unique())):
    for chrom in main_trnscpts_df["Chr"].unique():
        print("processing chromosome: chr", chrom)
        ### criar já o array do chromossoma correspondente à presença de exons
        exons = main_trnscpts_df.loc[(main_trnscpts_df["Type"] == "exon") & (main_trnscpts_df["Chr"] == str(chrom))]
        chr_exons_array = element_array(chromosomes_df, exons, chrom)
        ###
        chr_cds_dens, cds_exon_dens, cds_intron_dens = chr_process1(chromosomes_df, main_trnscpts_df, cytobands_df, chr_exons_array, chrom)
        #chr_utr5_dens, utr5_exon_dens, utr5_intron_dens = chr_process2(chromosomes_df, main_trnscpts_df, cytobands_df, chr_exons_array, chrom)
        chr_utr5_dens = chr_process2(chromosomes_df, main_trnscpts_df, cytobands_df, chr_exons_array, chrom)
        #chr_utr3_dens, utr3_exon_dens, utr3_intron_dens = chr_process3(chromosomes_df, main_trnscpts_df, cytobands_df, chr_exons_array, chrom)
        chr_utr3_dens = chr_process3(chromosomes_df, main_trnscpts_df, cytobands_df, chr_exons_array, chrom)
        chr_exon_dens = chr_process4(chromosomes_df, main_trnscpts_df, cytobands_df, chr_exons_array, chrom)
        ###
        all_chrs_cdsDens = all_chrs_cdsDens + chr_cds_dens
        all_chrs_cdsExonDens = all_chrs_cdsExonDens + cds_exon_dens
        all_chrs_cdsIntronDens = all_chrs_cdsIntronDens + cds_intron_dens
        all_chrs_utr5Dens = all_chrs_utr5Dens + chr_utr5_dens
        #all_chrs_utr5ExonDens = all_chrs_utr5ExonDens + utr5_exon_dens
        #all_chrs_utr5IntronDens = all_chrs_utr5IntronDens + utr5_intron_dens
        all_chrs_utr3Dens = all_chrs_utr3Dens + chr_utr3_dens
        #all_chrs_utr3ExonDens = all_chrs_utr3ExonDens + utr3_exon_dens
        #all_chrs_utr3IntronDens = all_chrs_utr3IntronDens + utr3_intron_dens
        all_chrs_exon_dens = all_chrs_exon_dens + chr_exon_dens
        #print("length of UTR5 densities so far: ", len(all_chrs_cdsExonDens))
    cytobands_df.insert(len(cytobands_df.columns), "CDS_dens", all_chrs_cdsDens, True)
    cytobands_df.insert(len(cytobands_df.columns), "CDS_exon_dens", all_chrs_cdsExonDens, True)
    cytobands_df.insert(len(cytobands_df.columns), "CDS_intron_dens", all_chrs_cdsIntronDens, True)
    cytobands_df.insert(len(cytobands_df.columns), "5P UTR_dens", all_chrs_utr5Dens, True)
    #cytobands_df.insert(len(cytobands_df.columns), "5P UTR_exon_dens", all_chrs_utr5ExonDens, True)
    #cytobands_df.insert(len(cytobands_df.columns), "5P UTR_intron_dens", all_chrs_utr5IntronDens, True)
    cytobands_df.insert(len(cytobands_df.columns), "3P UTR_dens", all_chrs_utr3Dens, True)
    #cytobands_df.insert(len(cytobands_df.columns), "3P UTR_exon_dens", all_chrs_utr3ExonDens, True)
    #cytobands_df.insert(len(cytobands_df.columns), "3P UTR_intron_dens", all_chrs_utr3IntronDens, True)
    cytobands_df.insert(len(cytobands_df.columns), "exon_dens", all_chrs_exon_dens, True)

```

```python
def writeNewtxt(file_path, df):
    """Writes a pandas dataframe into txt, columns separated by tabs ("\t")
    """
    f = open(file_path, 'w')
    df_cols = ""
    for col in df.columns[:-1]:
        df_cols += str(col) + "\t"
    df_cols += df.columns.values[-1]+ "\n"
    f.write(str(df_cols))
    for row in df.index:
        df_row = ""
        for col in df.columns[:-1]:
            df_row += str(df[col][row]) + "\t"
        df_row += str(df[df.columns[-1]][row])+ "\n"
        f.write(df_row)
    f.close
```
